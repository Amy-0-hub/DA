### RQ2: Which countries have produced the most Netflix content?

```python
#step1: load data
import pandas as pd
import matplotlib.pyplot as plt
country_data = pd.read_csv('/Users/yaqin/DA/netfilx_aws/data/countries_result.csv')  
country_data.head()
```

```python
#step2: clean data
country_data = country_data.dropna(subset=['country']) 
country_data['country'] = country_data['country'].astype(str).str.strip().str.replace('"', '')  

#step3: get top 3 countries
top_countries = country_data.sort_values(by='movie_count',ascending=False).head(4)   # missing one value 
#step3: Plot bar chart showing top countries by number of Netflix titles
plt.bar(top_countries['country'], top_countries['movie_count'], color='red', width=0.5)
plt.title('Top Countries by Netflix Content Production')
plt.xlabel('Countries')
plt.ylabel('Number of Titles')

plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```
