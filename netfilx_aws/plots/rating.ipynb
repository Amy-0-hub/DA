### RQ3: How have content ratings changed from 2014 to 2021?

```python
#step1: load data
import pandas as pd
import matplotlib.pyplot as plt
rating_data = pd.read_csv('/Users/yaqin/DA/netfilx_aws/data/rating_result.csv')  
rating_data.head()
```

```python
#step2: pivot table to convert lines
pivot_df = rating_data.pivot(index='release_year', columns='rating', values='title_count')
pivot_df = pivot_df.fillna(0)

#step3: plot a line chart, showing content rating changed from 2014 to 2021
for each_rating in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[each_rating], marker='o', label=each_rating)

plt.title('Netflix Content Ratings Over Time (2014-2021)')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Rating', loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Suggested Analysis:
    TV-MA dominates all other ratings by a significant margin, indicating a strong strategic shift toward adult-oriented content on Netflix.
