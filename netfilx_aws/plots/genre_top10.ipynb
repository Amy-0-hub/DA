### RQ1: What are the top 10 most common content genres on Netflix?

```python
#step1: load data
import pandas as pd
import matplotlib.pyplot as plt
genre_data = pd.read_csv('/Users/yaqin/DA/netfilx_aws/data/genres_result.csv')  
genre_data.head()
```

```python
#step2: plot a bar chart, showing top 10 most common genres on Netflix

plt.bar(genre_data['genre'], genre_data['count'], color='blue', width= 0.5)
plt.title('The Top 10 Most Common Content Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Count')

plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

```
